```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'png'
```

Define a function to calculate the Autocorrelation function of a 1D function varying over time, i.e.:
$$ C(f, m) = \dfrac{1}{N} \sum^{N-m-1}_{i=0} f_i \cdot f^*_{i+m} $$
This follows [this](http://jespertoftkristensen.com/JTK/Blog/Entries/2013/12/29_Efficient_computation_of_autocorrelations%3B_demonstration_in_MatLab_and_Python.html) tutorial.

```python
def acf_discrete(x):
    N = len(x)
    m = N-1
    acf = list()
    for lag in range(m):
        summ = 0
        for i in range(N-lag-1):
            summ += (x[i+1]*x[i+lag+1])
        acf.append(summ/N)
    return acf
```

Generate some periodic signal and compute the ACF as defined above:

```python
x = np.sin(np.linspace(0, np.pi*10, 1000))
plt.plot(acf_discrete(x))
```

This definition results in an ACF which decays over correlation time (due to finite data points overlap), and is not normalised.

### Some theory and tests

Function that calculates the serial correlation, i.e. the Pearson correlation matrix for the version of itself shifted by a specified *lag*. The Pearson Correlation Coefficient is defined as:

$$ \rho = \dfrac{\sum_i (x_i - \mu_x)(y_i-\mu_y)}{N\sigma_x \sigma_y} $$

Where $\mu$ are the means and $\sigma$ the standard deviations. It can have values between +1 (if one var is high the other is high too) and -1 (if one var is high the other is low).

```python
def serial_corr(x, lag=1):
    N = len(x)
    f1 = x[lag:]
    f2 = x[:N-lag] # in reality I am just shifting the limits of the two functions, 
                   # whilst keeping the n of points the same
    corr = np.corrcoef(f1,f2)[0,1] # I want the diagonal of the matrix
    return corr
```

Now, in principle in the case of a sine wave, the correlation coefficient a 0 lag should be +1, and at $\pi$ lag -1, because the two waves are completely out of phase. However, these are not continuous functions... but a range of points. So the values are not quite those.

```python
x = np.sin(np.linspace(0, np.pi*10, 1000)) # five periods, 1e3 points --> pi = 100 points.
lags = range(0,1000,100) # each lag = a period
for coeff in [serial_corr(x, lag=l) for l in lags]:
    print(coeff)
```

Now define a function the calculates the serial correlation, but for all allowed lags. 

```python
def autocorr(x):
    lags = range(len(x)//2)
    corrs = [serial_corr(x, lag) for lag in lags] # List comprehension!
    return lags, corrs
```

Note that the allowed shifts can only go as far as half of the total points in the supplied function, since I am shifting them with respect to eachother by simple indexing -- just selecting portions of the original one.

```python
plt.plot(*autocorr(x), label='acf(x)')
plt.plot(x, label='x')
plt.legend();
```

Note how $\text{acf}[\sin(x)] = \cos(x)$. 

Test on some random data:

```python
x = np.random.rand(1000)
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.plot(*autocorr(x))
plt.subplot(1,2,2)
plt.plot(*autocorr(x))
plt.xlim(0,50);
```

Test on sin function multiplied by random data:

```python
x = np.random.rand(1000)*np.sin(np.linspace(0,10*np.pi,1000))
plt.plot(*autocorr(x))
```

As expected for uniformely distributed random numbers over [0,1), the ACF is only 1 at its centre (point 0), and then oscillates around 0. So the random numbers are not correlated... obv. 

Notes on this way of calculating the autocorrelation:
- Always bounded between +1 and -1 by definition
- Does not decay for a periodic function; which makes sense.

Now using `np.correlate` and `np.convolve`. These accept two arrays of different shape. If shape1=N, shape2=M:

- `mode='full'` --> out shape (N+M-1). Convolution at each point of overlap. At the boundaries signals do not overlap completely, boundary effects *may* (?) be seen.
- `mode='same'` --> out shape max(M, N). Boundary effects still visible.
- `mode='valid'` --> out shape max(M,N) - min(M, N) + 1. Convolution/Corr **only** for points where the signal overlaps **completely**. Values outside the signal boundary have no effect.

```python
x = np.sin(np.linspace(0, np.pi*10, 1000))
```

```python
print(np.correlate(x,x,mode='valid'))
print(np.convolve(x,x,mode='valid'))
```

The two are equivalent, except for the "direction" in which the two functions are slided with respect to eachother in the convolution operation:

```python
print(np.correlate(x,x,mode='valid'))
print(np.convolve(x,np.flipud(x),mode='valid'))
```

See what happens with the other modes. Note the number of points in the two plots below!

```python
plt.plot(np.convolve(x,x,mode='full'), label='conv')
plt.plot(np.correlate(x,np.flipud(x),mode='full'), label='corr')
plt.legend();
```

```python
plt.plot(np.convolve(x,x,mode='same'), label='conv')
plt.plot(np.correlate(x,np.flipud(x),mode='same'), label='corr')
plt.legend();
```

"Normalise" and compare with user defined funcs

```python
plt.plot(np.convolve(x,np.flipud(x),mode='same')/len(x), label='conv')
plt.plot(np.correlate(x,x,mode='same')/len(x), label='corr')
plt.plot(*autocorr(x), label='pearson')
plt.plot(acf_discrete(x), label='edo_defined')
plt.legend();
```

Looks like I have reinvented the weel; my function and `np.correlate` actually produce the same result:

```python
plt.plot((np.correlate(x,x,mode='full')/len(x))[1000:], label='conv')
plt.plot(acf_discrete(x), label='edo')
plt.legend();
```

And again, more efficiently with `fftconvolve`:

```python
from scipy.signal import fftconvolve
```

```python
plt.plot(np.convolve(x,x,mode='same')/len(x), label='conv')
plt.plot(fftconvolve(x,x,mode='same')/len(x), label='fft')
plt.legend();
```

```python
plt.plot((fftconvolve(x,np.flipud(x),mode='full')/len(x))[1000:], label='fft')
plt.plot(acf_discrete(x), label='edo')
plt.legend();
```

Finally note the following:

```python
plt.plot(*autocorr(x), label='acf')
plt.plot((fftconvolve(x,np.flipud(x),mode='same')/len(x))[500:], label='fft_conv')
plt.legend();
```

The definition of ACF using the Pearson Correlation coeffient is the only one (as compared to the python implemented ones tested) that does not decay. But then obviously the user defind ACF can access only half of the data points. Test on random data:

```python
x = np.random.rand(1000)
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.plot(*autocorr(x), label='acf')
plt.plot((fftconvolve(x,x[::-1],mode='full')/len(x))[1000:], label='fft_conv')
plt.subplot(1,2,2)
plt.plot(*autocorr(x), label='acf')
plt.plot((fftconvolve(x,x[::-1],mode='full')/len(x))[1000:], label='fft_conv')
plt.xlim(0,5)
plt.legend()
```

Since `fftconvolve` (or any other `numpy` function of this sort) does not divide by the number of points at each product in the sum, and is not normalised, to compare the two the following is necessary:

```python
acf = fftconvolve(x,np.flipud(x),mode='same')
N = len(acf)
half_acf = acf[N//2:]
lengths = range(N, N//2, -1)
half_acf = half_acf/lengths
half_acf = half_acf / half_acf[0]


plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.plot(*autocorr(x), label='acf')
plt.plot(half_acf, label='fft_conv')
plt.legend()
plt.subplot(1,2,2)
plt.plot(*autocorr(x), label='acf')
plt.plot(half_acf, label='fft_conv')
plt.xlim(0,5)
plt.legend()
```

Why the hell does the `fftconvolve` result on top not oscillate around 0 as expected? It works correctly for normally distributed random numbers:

```python
x = np.random.randn(1000)
acf = fftconvolve(x,np.flipud(x),mode='same')
N = len(acf)
half_acf = acf[N//2:]
lengths = range(N, N//2, -1)
half_acf = half_acf/lengths
half_acf = half_acf / half_acf[0]


plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.plot(*autocorr(x), label='acf')
plt.plot(half_acf, 'o', label='fft_conv')
plt.legend()
plt.subplot(1,2,2)
plt.plot(*autocorr(x), label='acf')
plt.plot(half_acf, 'o', label='fft_conv')
plt.xlim(0,5)
plt.legend()
```

And periodic stuff (1-2% difference in values):

```python
x = np.sin(np.linspace(0, np.pi*10, 1000))
acf = fftconvolve(x,np.flipud(x),mode='same')
N = len(acf)
half_acf = acf[N//2:]
lengths = range(N, N//2, -1)
half_acf = half_acf/lengths
half_acf = half_acf / half_acf[0]


plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.plot(*autocorr(x), label='acf')
plt.plot(half_acf, 'o', label='fft_conv')
plt.legend()
plt.subplot(1,2,2)
plt.plot(*autocorr(x), label='acf')
plt.plot(half_acf, 'o', label='fft_conv')
plt.xlim(0,5)
plt.legend()
```

This is the same as `fftconvolve`:

```python
def acf_fft(f):
    
    N = len(f)
    fvi = np.fft.fft(f, n=2*N)
    acf =  fvi*np.conjugate(fvi)
    acf = np.fft.ifft(acf)[:N].real/N
    
    return acf
```
