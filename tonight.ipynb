```python
x = np.random.rand(10)
```

```python
def acf(x):
    N = len(x)
    acf = []
    x = x[:N/2]
    for m in range(N-1):
        R = 0
        for i in range(N-m):
            R += x[i]*x[i+m]
        acf.append(R)
    return np.array(acf)
```

```python
test = acf(x)
# test /= test[0]
```

```python
plt.plot(test,'o')
```
